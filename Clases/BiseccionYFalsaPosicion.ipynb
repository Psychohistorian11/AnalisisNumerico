```python
def biseccion(f,a,b,tol):
  if(f(a)*f(b)>0):
    print(f"la función no cumple el teorema en: {a,b}")
  else:
    i=0
    while(abs(b-a)>tol):
      c=(a+b)/2
      if (f(a)*f(c)<0):
        b=c
      else:
        a=c
      i+=1
    print("la solución es:",c,"y las iteraciones",i)
    return
```

```python
import numpy as np
f=lambda x:35*np.tan(x)-15.006/((np.cos(x))**2)+1
```

```python
biseccion(f,0,np.pi/4,1e-6)
```

```python
def falsaPosicion(f,a,b,tol):
  if(f(a)*f(b)>0):
    print(f"la función no cumple el teorema en: {a,b}")
  else:
    i=0
    c=a-f(a)*(a-b)/(f(a)-f(b))
    while(abs(f(c))>tol):
      c=a-f(a)*(a-b)/(f(a)-f(b))
      if (f(a)*f(c)<0):
        b=c
      else:
        a=c
      i+=1
    print("la solución es:",c,"y las iteraciones",i)
    return
```

```python
import numpy as np
f=lambda x:35*np.tan(x)-15.006/((np.cos(x))**2)+1
```

```python
falsaPosicion(f,0,np.pi/4,1e-6)
```
