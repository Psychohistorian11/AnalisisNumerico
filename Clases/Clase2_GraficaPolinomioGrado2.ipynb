```python
#Graficar una función y la construcción del polinomio de grado 2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1,100)
f = lambda x:2**(-x)
p = lambda x: 1-np.log(2)*x+(np.log(2))**2*x**2
plt.plot(x,f(x),'r', label='Función')
plt.plot(x,p(x),'b', label='Polinomio de grado 2')
plt.legend()
plt.xlabel ('x')
plt.ylabel('y')
plt.show()
```
