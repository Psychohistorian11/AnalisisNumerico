```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.01
to = 0
tf = 2
co= 30
k = -0.4056
Ta = 27
n = int((tf-to)/h)

T = [co] #co condición inicial

for i in range(0,n,1):
    T.append(T[i]-h*k*(T[i] - Ta))

for i in T:
  print(i, "\n")
```
